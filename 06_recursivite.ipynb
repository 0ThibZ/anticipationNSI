# Module 6 : La Puissance de la Récursivité 🪆

La récursivité est un concept magique : c'est une fonction qui s'appelle elle-même ! 
C'est comme une mise en abyme ou des poupées russes.

## 1. Exemple de base : Le Compte à rebours

```python
def compte_a_rebours(n):
    if n == 0:  # Le cas de base (pour s'arrêter !)
        print("Décollage 🚀")
    else:
        print(n)
        compte_a_rebours(n - 1)  # La fonction s'appelle elle-même !

compte_a_rebours(3)
```

3. Déplacer les $N-1$ disques de B vers C.
