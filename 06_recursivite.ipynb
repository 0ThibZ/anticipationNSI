## 🧩 Exercice Bonus : La mémoïsation — guérir Fibonacci (★★★)

Vous avez constaté que `fib(35)` est lent. Pourquoi ?
Parce que `fib(10)` est recalculé des **milliers de fois** inutilement.

La **mémoïsation** consiste à **mémoriser** les résultats déjà calculés dans un dictionnaire.
Python offre `@lru_cache` qui fait ça automatiquement en une seule ligne !

```python
from functools import lru_cache

@lru_cache(maxsize=None)   # Cette ligne suffit !
def fib_rapide(n):
    if n <= 1: return n
    return fib_rapide(n-1) + fib_rapide(n-2)
```

**À faire :**
1. Créez `fib_memo(n)` avec `@lru_cache` (recopiez simplement votre `fib` en ajoutant le décorateur).
2. Comparez le temps d'exécution de `fib(30)` vs `fib_memo(30)` avec le module `time`.
3. Vérifiez que `fib_memo(100)` répond instantanément — impossible sans mémoïsation !

````python
import time
from functools import lru_cache

# Version lente (votre fonction originale)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Version rapide avec mémoïsation
@lru_cache(maxsize=None)
def fib_memo(n):
    if n <= 1:
        return n
    return fib_memo(n-1) + fib_memo(n-2)

# Comparaison des temps :
n_test = 30

debut = time.time()
resultat_lent = fib(n_test)
temps_lent = time.time() - debut

debut = time.time()
resultat_rapide = fib_memo(n_test)
temps_rapide = time.time() - debut

print(f"fib({n_test})      = {resultat_lent}  en {temps_lent:.4f}s")
print(f"fib_memo({n_test}) = {resultat_rapide}  en {temps_rapide:.6f}s")
print(f"Accélération : {temps_lent / max(temps_rapide, 0.000001):.0f}x plus rapide !")

# La mémoïsation permet d'aller très loin :
print(f"fib_memo(100) = {fib_memo(100)}")
print(f"fib_memo(200) = {fib_memo(200)}")
````

````python
assert fib_memo(10) == 55
assert fib_memo(20) == 6765
assert fib_memo(50) == 12586269025, "fib_memo(50) doit répondre instantanément et correctement."

debut = time.time()
fib_memo(100)
duree = time.time() - debut
assert duree < 0.01, f"fib_memo(100) devrait être quasi-instantané, pas {duree:.3f}s"

print("✅ Exercice Bonus validé ! Une ligne de code (@lru_cache) change tout.")
print(f"Avec la mémoïsation, fib(100) = {fib_memo(100)}")
````
