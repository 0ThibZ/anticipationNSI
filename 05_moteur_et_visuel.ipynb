# Module 5 : Le Moteur Hanoï et la Magie Visuelle 🪄

Maintenant que vous avez créé votre classe `Pile`, nous allons l'utiliser pour construire l'objet principal de notre jeu : la classe `JeuHanoi`.
Cette classe contiendra 3 Piles (nos 3 piquets) et gérera les règles du jeu.

```python
class Pile:
    def __init__(self): self.elements = []
    def est_vide(self): return len(self.elements) == 0
    def empiler(self, v): self.elements.append(v)
    def depiler(self): return self.elements.pop() if not self.est_vide() else None
    def voir_sommet(self): return self.elements[-1] if not self.est_vide() else None

class JeuHanoi:
    def __init__(self, nb_disques):
        self.piquets = {'A': Pile(), 'B': Pile(), 'C': Pile()}
        # On place tous les disques sur le piquet A (du plus grand au plus petit)
        for d in range(nb_disques, 0, -1):
            self.piquets['A'].empiler(d)
            
    def est_mouvement_valide(self, depart, arrivee):
        pile_dep = self.piquets[depart]
        pile_arr = self.piquets[arrivee]
        
        if pile_dep.est_vide():
            return False
        if pile_arr.est_vide():
            return True
            
        return pile_dep.voir_sommet() < pile_arr.voir_sommet()
        
    def jouer_coup(self, depart, arrivee):
        if self.est_mouvement_valide(depart, arrivee):
            disque = self.piquets[depart].depiler()
            self.piquets[arrivee].empiler(disque)
            print(f"Déplacement du disque {disque} de {depart} vers {arrivee}")
        else:
            print("❌ Coup invalide !")
            
    def afficher(self):
        # Dans la version finale, cette méthode appellera matplotlib !
        print("A:", self.piquets['A'].elements)
        print("B:", self.piquets['B'].elements)
        print("C:", self.piquets['C'].elements)
        print("---")
```
