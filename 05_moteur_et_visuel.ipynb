# Module 5 : Le Moteur Hanoï et la Magie Visuelle 🪄

Maintenant que vous avez créé votre classe `Pile`, nous allons l'utiliser pour construire l'objet principal de notre jeu : la classe `JeuHanoi`.
Cette classe contiendra 3 Piles (nos 3 piquets) et gérera les règles du jeu.

```python
class Pile:
    def __init__(self): self.elements = []
    def est_vide(self): return len(self.elements) == 0
    def empiler(self, v): self.elements.append(v)
    def depiler(self): return self.elements.pop() if not self.est_vide() else None
    def voir_sommet(self): return self.elements[-1] if not self.est_vide() else None

class JeuHanoi:
    def __init__(self, nb_disques):
        self.piquets = {'A': Pile(), 'B': Pile(), 'C': Pile()}
        # On place tous les disques sur le piquet A (du plus grand au plus petit)
        for d in range(nb_disques, 0, -1):
            self.piquets['A'].empiler(d)
            
    def est_mouvement_valide(self, depart, arrivee):
        pile_dep = self.piquets[depart]
        pile_arr = self.piquets[arrivee]
        
        if pile_dep.est_vide():
            return False
        if pile_arr.est_vide():
            return True
            
        return pile_dep.voir_sommet() < pile_arr.voir_sommet()
        
    def jouer_coup(self, depart, arrivee):
        if self.est_mouvement_valide(depart, arrivee):
            disque = self.piquets[depart].depiler()
            self.piquets[arrivee].empiler(disque)
            print(f"Déplacement du disque {disque} de {depart} vers {arrivee}")
        else:
            print("❌ Coup invalide !")
            
    def afficher(self):
        # Dans la version finale, cette méthode appellera matplotlib !
        print("A:", self.piquets['A'].elements)
        print("B:", self.piquets['B'].elements)
        print("C:", self.piquets['C'].elements)
        print("---")
```

## 🧩 Exercice 3 : Visualisation graphique (★★★)

Remplaçons l'affichage texte par une vraie visualisation !
Complétez la méthode `afficher_graphique()` dans `JeuHanoiV4`.

L'idée : pour chaque piquet, dessiner des barres horizontales dont la **largeur** est proportionnelle à la taille du disque.
On utilisera `matplotlib`.

Le code ci-dessous est partiellement complété. Lisez-le et comprenez chaque ligne.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

class JeuHanoiV4(JeuHanoiV3):
    def afficher_graphique(self, titre="État du jeu"):
        fig, axes = plt.subplots(1, 3, figsize=(10, 5))
        noms = ['A', 'B', 'C']
        
        for i, nom in enumerate(noms):
            ax = axes[i]
            disques = self.piquets[nom].elements
            nb_disques_max = 5
            
            for j, disque in enumerate(disques):
                largeur = disque * 0.3
                x = (nb_disques_max * 0.3 - largeur) / 2
                rect = mpatches.FancyBboxPatch(
                    (x, j), largeur, 0.8,
                    boxstyle="round,pad=0.05",
                    color=plt.cm.RdYlGn(1 - disque / (nb_disques_max + 1))
                )
                ax.add_patch(rect)
            
            ax.set_xlim(0, nb_disques_max * 0.3)
            ax.set_ylim(0, nb_disques_max + 1)
            ax.set_title(f"Piquet {nom}", fontsize=14, fontweight='bold')
            ax.set_xticks([])
            ax.set_yticks([])
            ax.axvline(x=nb_disques_max * 0.3 / 2, color='brown', linewidth=3, zorder=0)
        
        plt.suptitle(titre, fontsize=16)
        plt.tight_layout()
        plt.show()

# Test :
jeu = JeuHanoiV4(4)
jeu.afficher_graphique("État initial — 4 disques sur A")
jeu.jouer_coup('A', 'C')
jeu.jouer_coup('A', 'B')
jeu.jouer_coup('C', 'B')
jeu.afficher_graphique(f"Après 3 coups ({jeu.nb_coups} coups joués)")
```

## 🧩 Exercice 4 : Partie complète automatique (★★★★)

Combinez tout ce que vous avez appris !

Créez une fonction `resoudre_et_visualiser(nb_disques)` qui :
1. Crée un `JeuHanoiV4` avec `nb_disques` disques
2. Calcule la séquence de coups optimale via l'algorithme récursif (fonction `hanoi_liste` ci-dessous)
3. Joue tous les coups sur le jeu
4. Affiche le résultat final avec `afficher_graphique()`
5. Affiche le nombre de coups joués

La fonction `hanoi_liste` est fournie — elle retourne la liste des coups.

```python
def hanoi_liste(n, source, cible, aux):
    """Retourne la liste des coups pour résoudre Hanoï."""
    if n == 1:
        return [(source, cible)]
    return (hanoi_liste(n-1, source, aux, cible) +
            [(source, cible)] +
            hanoi_liste(n-1, aux, cible, source))

def resoudre_et_visualiser(nb_disques):
    jeu = JeuHanoiV4(nb_disques)
    jeu.afficher_graphique(f"Début — {nb_disques} disques")
    coups = hanoi_liste(nb_disques, 'A', 'C', 'B')
    for source, dest in coups:
        jeu.jouer_coup(source, dest)
    print(f"Partie résolue en {jeu.nb_coups} coups (minimum théorique : {2**nb_disques - 1})")
    jeu.afficher_graphique(f"Fin — tous les disques sur C !")

resoudre_et_visualiser(4)
```
