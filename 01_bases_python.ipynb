## 3. Les Mathématiques 🧮

Python est aussi une excellente calculatrice. Vous pouvez utiliser `+` (addition), `-` (soustraction), `*` (multiplication), `/` (division), et `**` (puissance).

Un fait mathématique sur les Tours de Hanoï : le nombre minimum de coups pour résoudre le jeu avec $n$ disques est de $2^n - 1$.

```python
n = 3
coups_minimum = (2 ** n) - 1
print("Pour", n, "disques, il faut au minimum", coups_minimum, "coups pour gagner.")
```
