## 🧩 Exercice 2 : Valider un déplacement (★★☆)

Créez une fonction `est_deplacement_valide(source, dest)` qui prend deux listes (deux piquets) et retourne `True` si le déplacement est légal, `False` sinon.

Règles à respecter :
- Si `source` est vide → impossible de déplacer → `False`
- Si `dest` est vide → on peut toujours poser → `True`
- Sinon → valide seulement si le disque du sommet de `source` est **strictement plus petit** que celui du sommet de `dest`

````python
def est_deplacement_valide(source, dest):
    # Cas 1 : source vide
    if len(source) == 0:
        return False
    # Cas 2 : dest vide
    if len(dest) == 0:
        return True
    # Cas 3 : comparaison des sommets
    return source[-1] < dest[-1]

# Tests :
# est_deplacement_valide([3,2,1], [])       →  True
# est_deplacement_valide([], [3])            →  False
# est_deplacement_valide([3,2,1], [3,2])    →  True  (1 < 2, valide !)
````

````python
assert est_deplacement_valide([3, 2, 1], []) == True
assert est_deplacement_valide([], [3]) == False
assert est_deplacement_valide([3, 2, 1], [3, 2]) == True, "1 < 2, le coup est valide !"
assert est_deplacement_valide([2, 3], [1]) == False, "3 > 1, le coup est invalide !"
print("✅ Exercice 2 validé !")
````

## 🧩 Exercice 4 : Jouer une séquence de coups (★★★★)

Créez une fonction `jouer_sequence(A, B, C, mouvements)` qui :
1. Prend trois piquets (listes) et une liste de tuples `[('A', 'C'), ('A', 'B'), ...]`
2. Pour chaque mouvement, vérifie s'il est valide avec votre fonction `est_deplacement_valide`
3. Si valide : effectue le déplacement avec votre fonction `deplacer` et affiche le coup
4. Si invalide : affiche un message d'erreur et arrête (`return`)
5. Retourne le nombre de coups joués

Vous utiliserez les fonctions `deplacer` et `est_deplacement_valide` déjà codées.

````python
def jouer_sequence(A, B, C, mouvements):
    piquets = {'A': A, 'B': B, 'C': C}
    nb_coups = 0
    
    for source, dest in mouvements:
        pile_source = piquets[source]
        pile_dest = piquets[dest]
        
        if est_deplacement_valide(pile_source, pile_dest):
            deplacer(pile_source, pile_dest)
            nb_coups += 1
            print(f"Coup {nb_coups} : {source} → {dest}")
        else:
            print(f"❌ Coup invalide : {source} → {dest}. Arrêt.")
            return nb_coups
    
    return nb_coups

# Test avec la solution complète pour 3 disques :
A = [3, 2, 1]
B = []
C = []
solution = [('A','C'), ('A','B'), ('C','B'), ('A','C'), ('B','A'), ('B','C'), ('A','C')]
total = jouer_sequence(A, B, C, solution)
print(f"\nTotal : {total} coups. C = {C}")
````

## 🧩 Exercice Bonus : Représenter le jeu avec un dictionnaire (★★★)

Jusqu'ici, on avait trois variables séparées `A`, `B`, `C`. Ce n'est pas pratique : si on veut accéder à un piquet par son nom (une lettre), il faut un `if/elif/else`.

Un **dictionnaire** résout ça : c'est une structure qui associe des **clés** à des **valeurs**.

```python
jeu = {'A': [3, 2, 1], 'B': [], 'C': []}
print(jeu['A'])          # → [3, 2, 1]
jeu['B'].append(1)       # On peut modifier la valeur directement
```

**À faire :** Réécrivez la fonction `deplacer` pour qu'elle prenne un dictionnaire `jeu` et deux noms de piquets en paramètres, puis créez une fonction `afficher_jeu_dict(jeu)` qui affiche proprement les trois piquets.

````python
def deplacer_dict(jeu, source, destination):
    """Déplace le disque du sommet du piquet `source` vers `destination`."""
    if len(jeu[source]) == 0:
        print(f"❌ Le piquet {source} est vide !")
        return
    
    disque = jeu[source][-1]
    
    if len(jeu[destination]) > 0 and disque > jeu[destination][-1]:
        print(f"❌ Coup invalide : disque {disque} ne peut pas aller sur {jeu[destination][-1]}")
        return
    
    jeu[source].pop()
    jeu[destination].append(disque)
    print(f"Disque {disque} : {source} → {destination}")

def afficher_jeu_dict(jeu):
    for nom, disques in jeu.items():
        print(f"  {nom} : {disques}")
    print("---")

# Test :
jeu = {'A': [3, 2, 1], 'B': [], 'C': []}
afficher_jeu_dict(jeu)

deplacer_dict(jeu, 'A', 'C')
deplacer_dict(jeu, 'A', 'B')
deplacer_dict(jeu, 'C', 'B')
afficher_jeu_dict(jeu)
````

````python
jeu = {'A': [3, 2, 1], 'B': [], 'C': []}
deplacer_dict(jeu, 'A', 'C')
assert jeu == {'A': [3, 2], 'B': [], 'C': [1]}, f"Après A→C : {jeu}"
deplacer_dict(jeu, 'A', 'B')
assert jeu['B'] == [2], f"Après A→B : B devrait contenir [2], pas {jeu['B']}"
deplacer_dict(jeu, 'C', 'B')
assert jeu['B'] == [2, 1], f"Après C→B : B devrait contenir [2, 1]"
print("✅ Exercice Bonus validé ! Les dictionnaires rendent le code bien plus élégant.")
````
