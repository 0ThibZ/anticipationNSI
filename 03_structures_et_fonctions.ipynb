Pour représenter les 3 piquets de Hanoï, des variables simples ne suffisent plus. Il nous faut des **Listes**.

## 1. Les Listes
Une liste est une variable qui peut contenir plusieurs valeurs. En Python, on l'écrit avec des crochets `[]`.

```python
piquet_A = [3, 2, 1]  # Le plus grand disque (3) est en bas, le plus petit (1) en haut.
print("Le piquet A contient :", piquet_A)
```

On peut retirer le dernier élément (celui tout en haut) avec `.pop()`, et l'ajouter sur un autre piquet avec `.append()`.

```python
piquet_B = []
disque_pris = piquet_A.pop()  # On prend le 1
piquet_B.append(disque_pris)  # On le pose sur B

print("Piquet A :", piquet_A)
print("Piquet B :", piquet_B)
```

## 2. Les Fonctions
Une fonction est un bloc de code réutilisable. On la définit avec le mot-clé `def`.

```python
def saluer(nom):
    print("Bonjour", nom, "!")

saluer("Champion de Hanoï")
```
