# Module 2 : Logique conditionnelle et boucles 🔄

Pour qu'un programme soit intelligent, il doit pouvoir prendre des décisions. C'est le rôle des **conditions**.

## 1. Les Conditions (if, elif, else)

Aux Tours de Hanoï, il y a une règle d'or : **On ne peut jamais poser un grand disque sur un plus petit.**
Voyons comment apprendre cette règle à Python.

````python
disque_que_je_deplace = 3
disque_sur_lequel_je_pose = 5

if disque_que_je_deplace < disque_sur_lequel_je_pose:
    print("✅ Mouvement autorisé ! Le disque 3 est plus petit que le 5.")
else:
    print("❌ Mouvement interdit ! On ne pose pas un grand sur un petit.")
````

## 2. Les Booléens (Vrai ou Faux)

L'expression `3 < 5` vaut en réalité `True` (Vrai) en Python. Si c'était faux, elle vaudrait `False`.

## 🧩 Exercice 1 : Compte à rebours avec `while` (★☆☆)

La boucle `while` répète un bloc **tant qu'une condition est vraie**.
C'est différent du `for` qui compte un nombre fixe de fois.

```python
compteur = 3
while compteur > 0:
    print(compteur)
    compteur = compteur - 1  # IMPORTANT : on doit réduire le compteur !
print("Fin !")
```

## 🧩 Exercice 3 : Simuler une partie (★★★)

Voici la séquence optimale pour résoudre Hanoï avec 3 disques.
Chaque élément de la liste est un tuple `(source, destination)`.

Simulez cette séquence sur 3 listes représentant les piquets A, B, C.
À chaque coup, vérifiez que le disque du sommet de `source` est bien plus petit que le sommet de `dest` (ou que `dest` est vide).
Comptez le nombre de coups joués et arrêtez avec `break` si un coup est invalide.

````python
A = [3, 2, 1]   # Plus grand en bas, plus petit en haut
B = []
C = []

sequence = [('A','C'), ('A','B'), ('C','B'), ('A','C'), ('B','A'), ('B','C'), ('A','C')]

piquets = {'A': A, 'B': B, 'C': C}
nb_coups_joues = 0

for source, dest in sequence:
    pile_source = piquets[source]
    pile_dest   = piquets[dest]

    if len(pile_source) == 0:
        print(f"❌ Coup invalide : {source} est vide !")
        break

    disque = pile_source[-1]

    if len(pile_dest) == 0 or disque < pile_dest[-1]:
        pile_source.pop()
        pile_dest.append(disque)
        nb_coups_joues += 1
        print(f"Coup {nb_coups_joues} : disque {disque} de {source} → {dest}")
    else:
        print(f"❌ Coup invalide : disque {disque} ne peut pas aller sur {pile_dest[-1]}")
        break

print(f"\nPartie terminée en {nb_coups_joues} coups.")
print(f"C : {C}")
````

````python
assert nb_coups_joues == 7, f"La séquence complète fait 7 coups, pas {nb_coups_joues}."
assert C == [3, 2, 1], "Tous les disques devraient être sur C dans l'ordre [3, 2, 1] !"
print("✅ Exercice 3 validé ! Vous avez simulé une résolution complète de Hanoï.")
````

## 🧩 Exercice Bonus : Affichage ASCII des piquets (★★★)

Avec des **boucles imbriquées**, on peut dessiner ! L'objectif est d'afficher une représentation textuelle des disques sur un piquet.

Pour un disque de taille `k`, on dessine `k` fois le caractère `█`, centré sur une largeur fixe.

Exemple pour le piquet A avec les disques [3, 2, 1] (3 en bas, 1 en haut) :

```
    █    
   ██    
  ███    
--------- A
```

**À faire :** Complétez la fonction `afficher_piquet_ascii(disques, nom, largeur_max)`.
- `disques` : liste des disques (ex: `[3, 2, 1]`), le premier est le fond
- Affichez du **haut vers le bas** (donc le dernier de la liste en premier)
- Chaque disque de taille `k` s'affiche comme `k * "█"` centré sur `largeur_max` caractères
- Terminez par une ligne `"-" * largeur_max + "  " + nom`

````python
def afficher_piquet_ascii(disques, nom, largeur_max=9):
    # On affiche de haut en bas : on parcourt la liste à l'envers
    for disque in reversed(disques):
        barre = "█" * disque
        # center() centre le texte dans une largeur donnée
        print(barre.center(largeur_max))
    print("-" * largeur_max + "  " + nom)

def afficher_jeu_ascii(A, B, C, largeur_max=9):
    print()
    print("=== État du jeu ===")
    nb_lignes = max(len(A), len(B), len(C)) + 1
    colonnes = []
    for disques, nom in [(A, "A"), (B, "B"), (C, "C")]:
        lignes = []
        for _ in range(nb_lignes - 1 - len(disques)):
            lignes.append(" " * largeur_max)
        for disque in reversed(disques):
            lignes.append(("█" * disque).center(largeur_max))
        lignes.append("-" * largeur_max + " " + nom)
        colonnes.append(lignes)
    for i in range(nb_lignes):
        print("   ".join(col[i] for col in colonnes))

# Test :
afficher_jeu_ascii([3, 2, 1], [], [])
print()
afficher_jeu_ascii([3], [2], [1])
````

````python
import io, sys
captured = io.StringIO()
sys.stdout = captured
afficher_piquet_ascii([3, 2, 1], "A")
sys.stdout = sys.__stdout__
output = captured.getvalue()
assert "█" in output or "#" in output, "La fonction doit afficher les disques avec █ ou #."
assert "A" in output, "Le nom du piquet doit apparaître."
print("✅ Exercice Bonus validé ! Vous savez maintenant dessiner avec des boucles.")
````
