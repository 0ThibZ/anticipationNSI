## 1. La Classe `Pile` (Stack / LIFO)

Une Pile c'est comme une pile d'assiettes : le dernier élément posé est le premier qu'on retire (**L**ast **I**n, **F**irst **O**ut). C'est exactement le comportement d'un piquet de Hanoï.

```python
class Pile:
    def __init__(self):
        # À la création de l'objet, on initialise une liste vide cachée à l'intérieur
        self.elements = []
        
    def est_vide(self):
        return len(self.elements) == 0
    
    def empiler(self, disque):
        self.elements.append(disque)
        
    def depiler(self):
        if not self.est_vide():
            return self.elements.pop()
        return None
        
    def voir_sommet(self):
        if not self.est_vide():
            return self.elements[-1]
        return None
```

```python
# Testons notre nouvel objet !
mon_piquet = Pile()
mon_piquet.empiler(3)
mon_piquet.empiler(2)

print("Au sommet, il y a le disque :", mon_piquet.voir_sommet())
disque_retire = mon_piquet.depiler()
print("J'ai retiré le disque :", disque_retire)
```

## 🧩 Exercice 3 : Héritage — La classe `PiquetHanoi` (★★★)

L'**héritage** permet de créer une nouvelle classe qui **réutilise** tout ce qu'une classe parente a déjà.

Créez une classe `PiquetHanoi` qui hérite de `Pile` et ajoute :
- Un `nom` ("A", "B" ou "C") passé au constructeur `__init__`
- Une méthode `peut_recevoir(disque)` qui retourne `True` si on peut poser ce disque au sommet (respecte la règle de Hanoï)

```python
class PiquetHanoi(Pile):
    def __init__(self, nom):
        super().__init__()   # Appelle le __init__ de Pile (initialise self.elements)
        self.nom = nom
    
    def peut_recevoir(self, disque):
        # Rappel : on peut poser si le piquet est vide, OU si le disque est plus petit que le sommet
        if self.est_vide():
            return True
        return disque < self.voir_sommet()

# Test :
piquet_a = PiquetHanoi("A")
piquet_a.empiler(3)
piquet_a.empiler(2)
print(f"Piquet {piquet_a.nom} peut recevoir disque 1 ?", piquet_a.peut_recevoir(1))
print(f"Piquet {piquet_a.nom} peut recevoir disque 3 ?", piquet_a.peut_recevoir(3))
```

```python
p = PiquetHanoi("B")
assert p.peut_recevoir(5) == True, "Un piquet vide accepte n'importe quel disque."
p.empiler(3)
assert p.peut_recevoir(1) == True, "Disque 1 < sommet 3 : valide."
assert p.peut_recevoir(4) == False, "Disque 4 > sommet 3 : invalide."
assert p.nom == "B", "Le nom du piquet doit être conservé."
print("✅ Exercice 3 validé !")
```

## 🧩 Exercice 4 : Rendre la `File` robuste (★★★★)

Créez une classe `FileRobuste` qui hérite de `File` et améliore son comportement :

1. Ajoutez une méthode `est_vide()` qui retourne `True` si la file est vide.
2. Redéfinissez `defiler()` pour qu'elle retourne `None` si la file est vide (au lieu de planter avec une erreur).
3. Ajoutez une méthode `taille()` qui retourne le nombre d'éléments.

```python
class FileRobuste(File):
    def est_vide(self):
        return len(self.elements) == 0
    
    def taille(self):
        return len(self.elements)
    
    def defiler(self):
        if self.est_vide():
            return None
        return super().defiler()   # Appelle defiler() de la classe File

# Test :
f = FileRobuste()
print("Vide ?", f.est_vide())
f.enfiler("Alice")
f.enfiler("Bob")
f.enfiler("Charlie")
print(f"Taille : {f.taille()}")
print(f"Premier : {f.defiler()}")
print(f"Deuxième : {f.defiler()}")
print(f"Taille restante : {f.taille()}")
print(f"Défiler une file vide : {FileRobuste().defiler()}")
```

```python
f = FileRobuste()
assert f.est_vide() == True
assert f.defiler() is None, "Défiler une file vide doit retourner None."
f.enfiler(1); f.enfiler(2); f.enfiler(3)
assert f.taille() == 3
assert f.defiler() == 1, "FIFO : le premier entré doit sortir en premier."
assert f.taille() == 2
print("✅ Exercice 4 validé ! Votre File est maintenant robuste.")
```
